```python
import extruct
import numpy as np
import matplotlib.pyplot as plt
help(extruct)
```

```python
god = extruct.Multiverse()

pspec = extruct.PowerLaw()
bg = extruct.Background()
pspec.A = 1.3
uni1 = god.bang(1000, bg, pspec) 

print(uni1)
#m_den = extruct.DensityObs(100)

#god.measure(uni1, m_den)

denObs = extruct.DensityObs(50)
pkObs = extruct.PowerSpectrumObs(0,100,0,100 )
```

```python
for t in np.linspace(0,0.6,10):

    god.evolveAll(t)

    god.measure(uni1, pkObs)
    god.measure(uni1, denObs)

    pk = pkObs.getResult()
    den = denObs.getResult()

    
    plt.loglog(pk[0,:],pk[1,:])
    plt.show()
    plt.plot(den[0,:],den[1,:])
    plt.show()
    
```
