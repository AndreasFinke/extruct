```python
import extruct
import numpy as np
import matplotlib.pyplot as plt


```

```python
god = extruct.Multiverse()

pspec = extruct.PowerLaw()
bg = extruct.Background()
pspec.A = 0.00004

for i in range(1,200):
    god.bang(2000, bg, pspec) 
  
#m_den = extruct.DensityObs(100)

#god.measure(uni1, m_den)

denObs = extruct.DensityObs(50)
pkObs = extruct.PowerSpectrumObs(0,100,1 )
```

```python
for t in np.linspace(0,0.1,10):

    god.evolveAll(t)

    god.measureAll(pkObs)
    god.measureAll(denObs)

    pk = pkObs.getResult()
    den = denObs.getResult()

    
    plt.loglog(pk[0,:],pk[1,:])
    #plt.plot(pk[0,:],pk[1,:])
    plt.show()
    plt.plot(den[0,:],den[1,:])
    plt.show()
    
for t in np.linspace(0.1,100,100):

    god.evolveAll(t)

    god.measureAll(pkObs)
    god.measureAll(denObs)

    pk = pkObs.getResult()
    den = denObs.getResult()

    
    plt.loglog(pk[0,:],pk[1,:])
    #plt.plot(pk[0,:],pk[1,:])
    plt.show()
    plt.plot(den[0,:],den[1,:])
    plt.show()
    

```
